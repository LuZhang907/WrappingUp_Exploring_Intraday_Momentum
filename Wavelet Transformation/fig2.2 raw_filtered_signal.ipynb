```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv("/Users/luzhang/Documents/GitHub/WrappingUp_Exploring_Intraday_Momentum_python/Data/fig2.2_raw_filtered_signal.csv")
```

```python
df.head()
```

```python
df.columns[0]
```

```python
# rename the first column
df.columns = ['x','time','raw_signal','filtered_signal']
df.head()
```

```python
# data frame
df = pd.DataFrame(df)
# drop na
df = df.dropna()
#change sereis to datetime format
df['time']=pd.to_datetime(df['time'])
# set time as index
df = df.set_index('time')
df.head()
```

```python
# keep time
df['TIME'] = [d.time() for d in df.index]
df.head()
```

```python
# visualizing
import matplotlib.pyplot as plt
%matplotlib inline

fig, axes = plt.subplots(3, 1, figsize=(12, 12))

axes[0].plot(df.x, df.raw_signal)
axes[0].set_title("raw signal AAPL index 2008-01-02")



axes[1].plot(df.x, df.filtered_signal,color="orange")
axes[1].set_title("filtered signal AAPL index 2008-01-02")

axes[2].plot(df.x, df.raw_signal,  df.x, df.filtered_signal)
#axes[2].set_ylim([0, 60])
#axes[2].set_xlim([2, 5])
axes[2].set_title("raw & filtered signal AAPL index 2008-01-02");
```
