```python
import pandas as pd
import numpy as np
import statsmodels.regression.linear_model as sm
from sklearn.linear_model import LinearRegression
from statsmodels.tools.tools import add_constant
```

```python
# def a combination function which returns the direction combination of two input features
# imput parameters: val1, val2 
def combination(val1, val2):
    conditions = [((val1>0) & (val2 >0)), ((val1>0) & (val2 <0)), ((val1<0) & (val2 >0)), ((val1 <0 )& (val2 <0))]
    # 0 for ++, 1 for +-, 2 for -+, 3 for --
    choices = [0,1,2,3]
    # output the combinations
    return np.select(conditions, choices, default='NA')
```

```python
cat_cols = ['LH0_ON', 'LH0_FH', 'LH0_ONFH', 'LH0_M', 'LH0_SLH', 'LH0_ROD', 'ON_FH',
       'ON_ONFH', 'ON_M', 'ON_SLH', 'ON_ROD', 'FH_ONFH', 'FH_M', 'FH_SLH',
       'FH_ROD', 'ONFH_M', 'ONFH_SLH', 'ONFH_ROD', 'M_SLH', 'M_ROD',
       'SLH_ROD']
cat_data = pd.get_dummies(dat[cat_cols],drop_first = True)
```

```python
num_cols = ['R_LH', 'R_LH0', 'R_ROD', 'R_ON', 'R_FH', 'R_ONFH', 'R_M', 'R_SLH',]
data = pd.concat([dat[num_cols], cat_data], axis=1)
data = add_constant(data)
```

```python
y = data.R_LH
```

```python
X = data.drop(["R_LH","const"],axis=1)
```

```python
# define a function find the features after backward Elimination
def BackwardEli(X,y):
    cols = list(X.columns)
    pmax = 1
    while (len(cols)>0):
        p = []
        X_1 = X[cols]
        X_1 = add_constant(X_1)
        model = sm.OLS(y,X_1).fit()
        p = pd.Series(model.pvalues.values[1:],index=cols)
        pmax = max(p)
        feature_with_p_max = p.idxmax()
        if(pmax>0.05):
            cols.remove(feature_with_p_max)
        else:
            break
    return cols
```

```python
cols = BackwardEli(X,y)
final_input_X = add_constant(data[cols])
final = sm.OLS(y,final_input_X).fit()
final.summary()
```
