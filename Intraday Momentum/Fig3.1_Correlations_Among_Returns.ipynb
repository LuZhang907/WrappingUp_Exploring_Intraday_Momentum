```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# read data
AAPL = pd.read_csv("/Users/luzhang/Documents/GitHub/WrappingUp_Exploring_Intraday_Momentum/Data/AAPL_nscoreReturns.csv")
QQQ = pd.read_csv("/Users/luzhang/Documents/GitHub/WrappingUp_Exploring_Intraday_Momentum/Data/QQQ_nscoreReturns.csv")
SPY = pd.read_csv("/Users/luzhang/Documents/GitHub/WrappingUp_Exploring_Intraday_Momentum/Data/SPY_nscoreReturns.csv")
```

```python
AAPL = AAPL.drop(AAPL.columns[0],axis=1)
QQQ = QQQ.drop(QQQ.columns[0],axis=1)
SPY = SPY.drop(SPY.columns[0],axis=1)
```

```python
from matplotlib.gridspec import GridSpec
```

```python
fig = plt.figure(figsize=(18, 9))
gs = GridSpec(nrows=2, ncols=2)
ax0 = fig.add_subplot(gs[0, 0])
sns.heatmap(AAPL.corr(),ax = ax0,cmap='Blues',annot=True)
ax0.set_title("AAPL: Correlation Among Returns")
ax1 = fig.add_subplot(gs[1, 0])
sns.heatmap(QQQ.corr(),ax = ax1,cmap='Blues',annot=True)
ax1.set_title("QQQ: Correlation Among Returns")
ax2 = fig.add_subplot(gs[:, 1])
sns.heatmap(SPY.corr(),ax = ax2,cmap='Blues',annot=True)
ax2.set_title("SPY: Correlation Among Returns")
# save figures
# save figures
fig.savefig("/Users/luzhang/Documents/GitHub/WrappingUp_Exploring_Intraday_Momentum/Output Results/Fig3_1_ReturnCorr.pdf")
```
