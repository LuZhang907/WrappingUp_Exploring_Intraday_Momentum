```python
import pandas as pd
import numpy as np
import statsmodels.regression.linear_model as sm
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# def a combination function which returns the direction combination of two input features
# imput parameters: val1, val2 
def combination(val1, val2):
    conditions = [((val1>0) & (val2 >0)), ((val1>0) & (val2 <0)), ((val1<0) & (val2 >0)), ((val1 <0 )& (val2 <0))]
    # 0 for ++, 1 for +-, 2 for -+, 3 for --
    choices = [0,1,2,3]
    # output the combinations
    return np.select(conditions, choices, default='NA')
```

```python
dat = pd.read_csv("/Users/luzhang/Documents/GitHub/WrappingUp_Exploring_Intraday_Momentum_2rd_try/Data/GOOGL_nscoreReturns.csv")
```

```python
dat = dat.dropna()
```

```python
dat["LH0_ON"] = combination(dat["R_LH0"],dat["R_ON"])
dat["LH0_FH"] = combination(dat["R_LH0"],dat["R_FH"])
dat["LH0_ONFH"] = combination(dat["R_LH0"],dat["R_ONFH"])
dat["LH0_M"] = combination(dat["R_LH0"],dat["R_M"])
dat["LH0_SLH"] = combination(dat["R_LH0"],dat["R_SLH"])
dat["LH0_ROD"] = combination(dat["R_LH0"],dat["R_ROD"])

dat["ON_FH"] = combination(dat["R_ON"],dat["R_FH"])
dat["ON_ONFH"] = combination(dat["R_ON"],dat["R_ONFH"])
dat["ON_M"] = combination(dat["R_ON"],dat["R_M"])
dat["ON_SLH"] = combination(dat["R_ON"],dat["R_SLH"])
dat["ON_ROD"] = combination(dat["R_ON"],dat["R_ROD"])

dat["FH_ONFH"] = combination(dat["R_FH"],dat["R_ONFH"])
dat["FH_M"] = combination(dat["R_FH"],dat["R_M"])
dat["FH_SLH"] = combination(dat["R_FH"],dat["R_SLH"])
dat["FH_ROD"] = combination(dat["R_FH"],dat["R_ROD"])

dat["ONFH_M"] = combination(dat["R_ONFH"],dat["R_M"])
dat["ONFH_SLH"] = combination(dat["R_ONFH"],dat["R_SLH"])
dat["ONFH_ROD"] = combination(dat["R_ONFH"],dat["R_ROD"])

dat["M_SLH"] = combination(dat["R_M"],dat["R_SLH"])
dat["M_ROD"] = combination(dat["R_M"],dat["R_ROD"])

dat["SLH_ROD"] = combination(dat["R_SLH"],dat["R_ROD"])
```

```python
dat.to_csv("/Users/luzhang/Documents/GitHub/WrappingUp_Exploring_Intraday_Momentum_2rd_try/Data/GOOGL_uscoreReturns_CTR.csv")
```

```python
cat_cols = ['LH0_ON', 'LH0_FH', 'LH0_ONFH', 'LH0_M', 'LH0_SLH', 'LH0_ROD', 'ON_FH',
       'ON_ONFH', 'ON_M', 'ON_SLH', 'ON_ROD', 'FH_ONFH', 'FH_M', 'FH_SLH',
       'FH_ROD', 'ONFH_M', 'ONFH_SLH', 'ONFH_ROD', 'M_SLH', 'M_ROD',
       'SLH_ROD']
cat_data = pd.get_dummies(dat[cat_cols])
```

```python
num_cols = ['R_LH', 'R_LH0', 'R_ROD', 'R_ON', 'R_FH', 'R_ONFH', 'R_M', 'R_SLH',]
data = pd.concat([dat[num_cols], cat_data], axis=1)
```

```python
y = data.R_LH
```

```python
X = data.drop("R_LH",axis=1)
```

```python
lr_full = sm.OLS(y,X).fit()
```

```python
lr_full.summary()
```

```python
pvalues = pd.DataFrame(lr_full.pvalues)
pvalues.columns = ["p_values"]
params = pd.DataFrame(lr_full.params)
params.columns = ["coefficents"]
table = pd.concat([params,pvalues],axis=1).sort_values("p_values")
```
